2) On observe que les courbes sont tangentes en les points d'abscisses $\pm\pi$. Tracer les tangentes communes aux deux courbes.



```python
f(x) = cosh(x) * cos(x)
g(x) = -cosh(x)
coef = diff(g)(pi)
t1(x) = -coef * x - coef * pi + g(-pi)
t2(x) = coef * x - coef * pi + g(-pi)
plot([f, g, t1, t2], (-5, 5))
```

### Exercice n° 4 : Résolution graphique



Résoudre les inéquations suivantes

$$\begin{array}{l}\frac{2 \, x^{2} + 5 \, x + 3}{x^{2} + x - 2}\geq 0 \\{\left| 4 \, x + 5 \right|} \leq 2\\\frac{x}{2 - x} \leq 1\\{\left| -3 \, x^{2} + x - 1 \right|} \geq 5\\x + 1 < \sqrt{x + 4}\\x - 3 \geq \sqrt{x^{2} - 5 \, x}\\ \sqrt{-x + 4} -\sqrt{2 \, x + 5}\geq 0\end{array}$$



```python
f(x) = (2 * x ** 2 + 5 * x + 3) / (x ** 2 + x - 2) # S ≃ ]-∞ ; -2] U [-5 / 3 ; -1] U [1 ; +∞[
g(x) = abs(4 * x + 5)                              # S ≃ [-2 ; -2 / 3]
h(x) = x / (2 - x)                                 # S ≃ ]-∞ ; 1] U ]2 ; +∞[
j(x) = abs(-3 * x ** 2 + x - 1)                    # S ≃ ]-∞ ; -1] U [4 / 3 ; +∞[
k(x) = x + 1 - sqrt(x + 4)                         # S ≃ [-4 ; 5 / 8[
l(x) = x - 3 - sqrt(x ** 2 - 5 * x)                # S ≃ [5 ; 19 / 2]
m(x) = sqrt(-x + 4) - sqrt(2 * x + 5)              # S ≃ [-5 / 2 ; -1 / 5]

show(plot([f, g, h, j ,k ,l, m], (-10, 10), ymin = -10, ymax = 10))
```

### Exercice n° 6 : Fonction intermédiaire



On appelle nombre intermédiaire entre trois nombres $a$, $b$ et $c$ le nombre qui est encadré par les deux autres. Par exemple, le nombre intermédiaire entre 42, 81 et 12 est 42. De même, le nombre intermédiaire entre 42, 42 et 81 est 42 car $42\leq 42\leq 81$.

Etant donné trois fonctions numériques `f`, `g` et `h`, écrire une fonction Python `intermed(x)` qui renvoie le nombre parmi `f(x)`, `g(x)` et `h(x)` et qui est situé entre les deux autres. On demande d'écrire une fonction Python avec `def` et pas une fonction formelle.

Pour le code de la fonction `intermed(x)`, on utilisera une liste que l'on triera avec la fonction `sorted`. Voici un exemple d'utilisation de `sorted`  :



```python
L=[42, 81, 12]
M=sorted(L)
print(L)
print(M)
```

Etant donné trois fonctions formelles f, g et h, et la fonction `intermed` définie ci-dessus, construire les graphes des fonctions f, g et h et en traçant en noir et en épais la fonction `intermed`.

Tester pour $f(x)=\ln(x^2)\sin(x)$, $g(x)=\sin(x^2)$ et $h(x)=x^2-1/2$. On obtiendra le dessin suivant :

![](interm.png)



```python
f(x) = ln(x ** 2) * sin(x)
g(x) = sin(x ** 2)
h(x) = x ** 2 - 1 / 2

def intermed(x):
    l = [f(x), g(x), h(x)]
    l.sort()
    return l[1]

show(plot([f, g, h], (-10, 10)))
show(plot(intermed, (-10, 10)))
```
